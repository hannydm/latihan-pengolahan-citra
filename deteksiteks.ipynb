```python
import cv2
import pytesseract
from google.colab.patches import cv2_imshow #nampilin gambar
```

```python
#Baca gambar
image = cv2.imread("test deteksi teks.jpg")
```

```python
#Konversi ke grayscale (biar OCR lebih akurat)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
#proses thresholding (naikin contrast teks)
gray = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]
```

```python
#nyimpen hasil preprocessing (buat ngecek)
cv2.imwrite("processed_image.jpg", gray)
```

```python
#ekstrak teks dari gambar
text = pytesseract.image_to_string(gray)
```

```python
#nampilin gambar yang di prosess
cv2_imshow(gray)w
cv2.waitKey(0)
cv2.destroyAllWindows()


```

```python
#output hasil baca gambar jadi teks
print("Hasil baca:", text)
```
